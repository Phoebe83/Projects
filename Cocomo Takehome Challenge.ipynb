In this analysis, we aim to evaluate the impact of various digital marketing channels on weekly revenue to assess the efficiency of current budget allocation across channels. The dataset includes weekly spend and impressions across key marketing channels (Google, Facebook, and Affiliates) along with associated revenue data, broken down by geography. The goal is to determine the channels that most significantly influence revenue and to understand whether the current allocation of marketing spend is efficient, as well as to see if there are differences across geo locations .

```python
# Load the dataset
df = pd.read_csv('/drive/data/media_data.csv')
```

#### II. Regression Modeling:

```python
# Define dependent and independent variables
# We remove spend here because they are highly correlated with the impression
X = df[['GOOGLE_IMP', 'FB_IMP', 'AFF_IMP', 'EMAIL_IMP', 'YT_PAID_IMP', 'YT_ORG_IMP']]
y = df['REVENUE']

# Add constant for intercept
X = sm.add_constant(X)

# Fit the OLS (Ordinary Least Squares) regression model
model = sm.OLS(y, X).fit()

# Summary of the model
print(model.summary())

# Get the p-values
p_values = model.pvalues
print("\nP-values:\n", p_values)
```

Model Performance: <br>
* R-squared (0.772): The model explains 77.2% of the variance in revenue. This indicates a strong fit, as a large proportion of the variability in revenue is explained by the independent variables (impressions from Google, Facebook, Affiliates, Email, YouTube Paid, and YouTube Organic).
* Adjusted R-squared (0.772): The adjusted R-squared value is identical to the R-squared, showing the model performs well even after adjusting for the number of predictors.
* F-statistic (1720, p < 0.001): The model is statistically significant overall, with a very low p-value (0.00). This means that the independent variables collectively have a significant relationship with the dependent variable (revenue).

Coefficients and Their Interpretation: <br>
* GOOGLE_IMP (0.2792, p < 0.001): Google impressions have a strong and statistically significant positive impact on revenue. For every unit increase in Google impressions, revenue increases by 0.2792 units.

* FB_IMP (0.3382, p < 0.001): Facebook impressions also have a significant and strong positive impact on revenue. For each unit increase in Facebook impressions, revenue increases by 0.3382 units, making Facebook impressions the most impactful predictor in this model.

* AFF_IMP (-0.0595, p < 0.001): Affiliate impressions have a significant negative effect on revenue. For each unit increase in affiliate impressions, revenue decreases by 0.0595 units. This suggests that affiliate impressions may be inefficient or even detrimental to revenue generation.

* EMAIL_IMP (0.4379, p < 0.001): Email impressions have the highest positive impact on revenue. Each unit increase in email impressions results in a 0.4379 unit increase in revenue, confirming the strength of email marketing as a revenue driver.

* YT_PAID_IMP (-0.0395, p = 0.006): Paid YouTube impressions negatively affect revenue in a statistically significant way. For each unit increase in paid YouTube impressions, revenue decreases by 0.0395 units. This suggests inefficiencies in paid YouTube marketing.

* YT_ORG_IMP (0.0276, p = 0.070): Organic YouTube impressions have a small positive effect on revenue, though the result is not statistically significant at the 5% level (p = 0.07). The impact of organic YouTube impressions is weaker compared to other channels.

#### II. Regression Analysis Across Region

1. We can add the geo in our model with one hot encoding for geo. 

```python
data = pd.get_dummies(df, columns=['GEO'], drop_first=True)  # One-hot encoding for geo. Drop one of the columns to avoid perfect multicollinearity
data = data.drop(columns=['WEEK_START', 'GOOGLE_SPEND', 'FB_SPEND', 'AFF_SPEND']).astype(int)

# Define features and target
X = data.drop('REVENUE', axis=1)  # Features
y = data['REVENUE']  # Target variable

# Add constant for intercept
X = sm.add_constant(X)

# Fit the OLS (Ordinary Least Squares) regression model
model = sm.OLS(y, X).fit()

# Summary of the model
print(model.summary())

# Get the p-values
p_values = model.pvalues
print("\nP-values:\n", p_values)
```

```python
# Use the Variance Inflation Factor (VIF) to identify which features are most affected by multicollinearity.  
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

The results indicate a multicollinearity issue. VIF analysis shows that GEO_B has a VIF value above 5, but the issue persists even after removing GEO_B. Further investigation could be conducted if more time were available. <br>

2. We can also incorporates interaction terms between GEO and the marketing channels:

```python
# Specify GEO as a categorical variable and create interaction terms with the marketing channels
# This will automatically create dummy variables for GEO and interaction terms with each channel.
# Define the model formula to include GEO, the main marketing channels, and interaction terms.
formula = """REVENUE ~ GOOGLE_IMP * GEO + FB_IMP * GEO + AFF_IMP * GEO + EMAIL_IMP * GEO + 
             YT_PAID_IMP * GEO + YT_ORG_IMP * GEO"""

# Using patsy to create design matrices, which statsmodels can use
y, X = dmatrices(formula, data=df, return_type='dataframe')

# Standardize independent variables (excluding GEO dummy variables and interaction terms)
scaler = StandardScaler()
X[X.columns.difference(['Intercept'])] = scaler.fit_transform(X[X.columns.difference(['Intercept'])])

# Fit the OLS model with interaction terms
model = sm.OLS(y, X).fit()

# Display the summary of the model results
print(model.summary())
```

Model Fit: <br>
R-squared (0.827) and Adj. R-squared (0.816) indicate that approximately 82.7% of the variation in REVENUE is explained by the model's predictors.
The F-statistic (75.88, p < 0.001) suggests the model is statistically significant overall, so the predictors do add value to explaining revenue variation.

Coefficients: <br>
Geographical Regions:
Regions B and E show a statistically significant impact on REVENUE. For example, GEO[T.B] has a positive effect (coef = 1.065e+05, p < 0.001), indicating that REVENUE is expected to increase by ~106,500 units in Region B.
Other regions are not significant (p > 0.05), meaning they don't contribute much explanatory power for REVENUE.

Advertising Impressions: <br>
GOOGLE_IMP has a positive but non-significant effect on REVENUE (coef = 6.766e+04, p > 0.05), while FB_IMP shows a positive and significant effect (coef = 6.595e+04, p < 0.05), suggesting that Facebook ads drive significant revenue.

Interactions: <br>
Several interaction terms (GOOGLE_IMP:GEO and FB_IMP:GEO) explore if the impact of ad impressions varies across regions. However, most of these interactions are non-significant, meaning advertising effects do not differ substantially by region.

Overall:<br>
The model suggests that REVENUE is most influenced by Region B and E and Facebook ad impressions, with other variables contributing less explanatory power or showing insignificant effects.

Examining specific periods of higher revenue might reveal insights into effective marketing or external events influencing customer engagement.

* Channel Spend Trends:
Spending patterns across Google, Facebook, and Affiliate channels vary, with certain periods showing concentrated or increased investments in one channel over the others.
Google appears to have the most stable spending trend, with relatively consistent levels over time. This stability could imply a foundational role in the marketing strategy, perhaps aimed at building ongoing engagement.

Affiliate spending also varies but with fewer and more pronounced peaks, which may reflect tactical partnerships or promotions. The concentrated spikes could imply that Affiliates are used strategically rather than as a constant source of traffic.

* Trends and Effectiveness Over Time:
If we overlay or correlate the revenue trends with channel spends, we might see that certain channels or combinations yield higher revenue growth at specific times.
For example, if revenue shows a notable increase after a spike in Facebook spend, there may be a lagged response to Facebook campaigns that should be factored into future spending strategies.

* Non-Linear Spending Patterns:
Given the fluctuations in spending, testing for non-linear relationships between spend and revenue (e.g., diminishing returns) is also valuable. If increased spending in a channel like Google or Facebook does not correspond to proportional revenue growth, incorporating a quadratic term in the regression model can capture these diminishing returns.
Alternatively, a piecewise regression could help if there is a threshold beyond which additional spend is less effective.


* Recommendations for Strategy Adjustments:
If seasonality is detected, a cyclical budget approach might be advantageous, focusing resources during high-impact periods and pulling back during low seasons.
A multi-channel strategy leveraging interactions between Google, Facebook, and Affiliates could potentially amplify returns by reaching customers in diverse ways. For instance, retargeting customers acquired via Google on Facebook might sustain engagement and enhance conversion.
